#                   Algorithms
An algorithm is a finate sequence of instructions to solve a problems or to perform a computation.
The selection of the correct algorithm is critical, the main issue is of course picking or designing a algorithm that does the job correctly, but it's also important to do the job in a reasonable amount of time.
The time scaling of algorithms is described with Big O notation.
where n is the number of datapoints in the input and O (n) is the number of steps needed to run the algorithm.

Here are a few examples of algorithms and there time complexety 


| Notation | Name   | Example  |
| --- | --- | --- |
|   O(1)  | Constant| adding an element to a unsorted list  |
|   O(log n)  | logarithmic | Finding a item in a sorted list  |
|   O( n)  | linear | Calculating an average  |
|   O( n*log n)  | linear | Sorting a list  |
|   O( n^2)  | quadratic | Multiplyingtwo n-digit numbers by a simple algorithm. Creating DOT-plots  |
|   O( n!)  | factorial | Traveling Salesman (brute force)  |
|   O( 2^n)  | exponential | common in many AI algorithms  |

Sorting a a classical example to describe algorithms.

[here](https://macr.ae/article/sorting-algorithms) is a link to visual explanation of diffrent sorting algorithms

```python
#######################################
#-Loading python modules
######################################
from matplotlib import pyplot as plt
from scipy.misc import factorial
import numpy as np
import math
#-------------------------------------


     

#########################################
#-Creating the data
#########################################
n           = np.arange(1,101,1) # Creating a array parmeters are (Start,Stop,Step)
log         = np.log2(n)         # Creating a array that is derived from array n
loglinear   = np.log2(n)*n        
quadratic   = n**2                
exponential = 2**n           
#-----------------------------------------



#######################################################
#-Setting up the graph below
#######################################################
%matplotlib inline    
fig=plt.figure()                 # Creating a image  
image =fig.add_axes([0,0,2,2])   # setting size 
image.set_title("Big O notation")
image.set_xlabel("Items")
image.set_ylabel("Steps")

#image.legend(("log(n)",'N', 'N*log(n)', 'N*N',"Exp"), loc=(1.05, 0.5))  

#--------------------------------------------------


###############################
#-Plotting curves
###############################
image.plot(n,log)                 
image.plot(n,n)                   
image.plot(n,loglinear)          
image.plot(n,quadratic)             
#image.plot(n,exponential)

#---------------------------------------

image.legend(("log(n)",'N', 'N*log(n)', 'N*N'), loc=(1.05, 0.5)) 
 
```
